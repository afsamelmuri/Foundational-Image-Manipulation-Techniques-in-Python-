## 1. Reading, Displaying, and Saving an Image

This task reads an image, displays it inline, and saves it.

```python

import cv2
from matplotlib import pyplot as plt

# Read the image
image = cv2.imread('story_lena_lenna_1.jpg')  # Replace with the correct path if needed
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for matplotlib display

# Display the original image
plt.imshow(image_rgb)
plt.axis('off')
plt.title('Original Image')
plt.show()

# Save the image
cv2.imwrite('saved_image.jpg', image)  # Saves the image as 'saved_image.jpg'
```

## 2. Converting and Displaying Images: RGB to Grayscale

This section converts an RGB image to grayscale and displays the comparison.

```python

# Read the image
image = cv2.imread('SweetPeppers.jpg')  # Replace with the correct path if needed
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for matplotlib display
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display original and grayscale images side by side
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image_rgb)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(gray_image, cmap='gray')
ax[1].set_title('Grayscale Image')
ax[1].axis('off')
plt.show()
```

## 3. Modifying an Image: Changing the Center Pixel Color

In this part, the center pixel of the image is changed to red, shown alongside the original image.

```python

# Read the image
image = cv2.imread('cameraman.png')  # Replace with the correct path if needed
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
height, width, _ = image.shape

# Calculate the center coordinates
center_y, center_x = height // 2, width // 2

# Change the center pixel to red
modified_image = image.copy()
modified_image[center_y, center_x] = [0, 0, 255]  # Red in BGR
modified_image_rgb = cv2.cvtColor(modified_image, cv2.COLOR_BGR2RGB)

# Display the original and modified images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image_rgb)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(modified_image_rgb)
ax[1].set_title('Center Pixel Modified')
ax[1].axis('off')
plt.show()
```

## 4. Adding Text to an Image

This section adds text to the image and displays both the original and modified versions.

```python

# Read the image
image = cv2.imread('Beach.jpg')  # Replace with the correct path if needed
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Add text to the image
modified_image = image.copy()
cv2.putText(modified_image, 'OpenCV Text', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
modified_image_rgb = cv2.cvtColor(modified_image, cv2.COLOR_BGR2RGB)

# Display original and text-modified images side by side
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image_rgb)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(modified_image_rgb)
ax[1].set_title('Image with Text')
ax[1].axis('off')
plt.show()
```

## 5. Modifying a Region of an Image: Changing a 20x20 Pixel Area to Blue

This section modifies a 20x20 pixel region to blue and displays both versions for comparison.

```python

# Read the image
image = cv2.imread('Baboon.png')  # Replace with the correct path if needed
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Modify a 20x20 region to blue
modified_image = image.copy()
start_x, start_y = 50, 50
modified_image[start_y:start_y+20, start_x:start_x+20] = [255, 0, 0]  # Blue in BGR
modified_image_rgb = cv2.cvtColor(modified_image, cv2.COLOR_BGR2RGB)

# Display original and modified images side by side
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image_rgb)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(modified_image_rgb)
ax[1].set_title('Region Modified to Blue')
ax[1].axis('off')
plt.show()
```

## 6. Converting RGB Images to CMYK and YCrCb Color Spaces

This part converts the image to CMYK (PIL) and YCrCb (OpenCV) color spaces and displays both.

```python

from PIL import Image

# Read the image and convert to CMYK using PIL
image_pil = Image.open('mandril.jfif')  # Replace with the correct path if needed
cmyk_image = image_pil.convert('CMYK')

# Convert to RGB for display
image_rgb = cv2.cvtColor(cv2.imread('mandril.jfif'), cv2.COLOR_BGR2RGB)
ycrcb_image = cv2.cvtColor(cv2.imread('mandril.jfif'), cv2.COLOR_BGR2YCrCb)

# Display RGB, CMYK, and YCrCb images
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(image_rgb)
ax[0].set_title('Original Image (RGB)')
ax[0].axis('off')

# Convert CMYK to RGB for display in matplotlib
cmyk_display = cmyk_image.convert("RGB")
ax[1].imshow(cmyk_display)
ax[1].set_title('CMYK Image')
ax[1].axis('off')

ax[2].imshow(ycrcb_image)
ax[2].set_title('YCrCb Image')
ax[2].axis('off')
plt.show()
```
